```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
prefix = '15000_epochs_pen003_'
```

```python
PAR = np.load(prefix + 'PAR_list.npy')
```

```python
def get_agg(obj, sum_over):
    obj_sum = 0
    obj_list = []
    sum_over = sum_over
    for ii, item in enumerate(obj):
        obj_sum += item
        if ii % sum_over == 0:
            obj_list.append(obj_sum / float(sum_over))
            obj_sum = 0
    obj_agg = np.array(obj_list)
    return obj_agg
```

```python
PAR_agg = get_agg(PAR, 100)
```

```python
plt.plot(PAR_agg[1:]);
```
