```python
import numpy as np
import matplotlib.pyplot as plt
from pylab import rcParams
rcParams['figure.figsize'] = 15, 7
```

```python
pen = 0.8
prefix = '200000_epochs_pen' + str(pen) + '_scale1000_'
```

```python
average_over = 100
start_index = 100000
end_index = 120000
```

```python
PAR = np.load(prefix + 'PAR_list.npy')
```

```python
def get_agg(obj, sum_over):
    obj_sum = 0
    obj_list = []
    sum_over = sum_over
    for ii, item in enumerate(obj):
        obj_sum += item
        if ii % sum_over == 0:
            obj_list.append(obj_sum / float(sum_over))
            obj_sum = 0
    obj_agg = np.array(obj_list)
    return obj_agg
```

```python
PAR_agg = get_agg(PAR[start_index:end_index], average_over)
```

```python
plt.plot(PAR_agg[1:]);
plt.xlabel('Epochs x 100', fontsize = 24);
plt.xticks(fontsize = 14);
plt.ylabel('PAR',fontsize = 24);
plt.yticks(fontsize = 14);
```

```python
np.save('final_presentation/PARs/' + str(pen) + '.npy', PAR_agg)
```
