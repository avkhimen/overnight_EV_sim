```python
prefix = '1000000_epochs_pen0.1_scale1000_'
```

```python
average_over = 1
```

```python
PAR = np.load(prefix + 'PAR_list.npy')
```

```python
def get_agg(obj, sum_over):
    obj_sum = 0
    obj_list = []
    sum_over = sum_over
    for ii, item in enumerate(obj):
        obj_sum += item
        if ii % sum_over == 0:
            obj_list.append(obj_sum / float(sum_over))
            obj_sum = 0
    obj_agg = np.array(obj_list)
    return obj_agg
```

```python
PAR_agg = get_agg(PAR[:40000], average_over)
```

```python
plt.plot(PAR_agg[1:]);
```
